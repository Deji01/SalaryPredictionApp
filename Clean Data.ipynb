```python
col_names = ['Age Group', 'Gender', 'Country', 'Education Level', 'Recent Job Title', 'Coding Experience', 'Python', 'R', 'SQL', 'C', 'C++', 'Java', 'Javascript', 'Julia','Swift', 'Bash', 'MATLAB','Industry', 'Company Size', 'Data Science Employees', 'Annual Salary ($USD)']
```

```python
selected_columns = [
'What is your age (# years)?',
'What is your gender? - Selected Choice',
'In which country do you currently reside?',
'What is the highest level of formal education that you have attained or plan to attain within the next 2 years?',
'Select the title most similar to your current role (or most recent title if retired): - Selected Choice',
'For how many years have you been writing code and/or programming?',
'What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - Python',
'What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - R',
'What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - SQL',
'What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - C',
'What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - C++',
'What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - Java',
'What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - Javascript',
'What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - Julia',
'What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - Swift',
'What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - Bash',
'What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - MATLAB',
'In what industry is your current employer/contract (or your most recent employer if retired)? - Selected Choice',
'What is the size of the company where you are employed?',
'Approximately how many individuals are responsible for data science workloads at your place of business?',
'What is your current yearly compensation (approximate $USD)?',
]
```

```python
len(col_names), len(selected_columns)
```

```python
col_dict = dict(zip(selected_columns, col_names))
pprint(col_dict, indent=10, width=200)
```

```python
df_to_clean = df2[selected_columns]
df_to_clean.head(2)
```

```python
df_to_clean.rename(columns=col_dict, inplace=True)
df_to_clean.head()
```

```python
df_to_clean.iloc[ : , 6:17] = np.where(df_to_clean.iloc[ : , 6:17].isna(), '0', '1')
```

```python
df_to_clean.head()
```

```python
variables = pd.DataFrame(columns = ['variable','dtype', 'nunique', 'unique'])

for var in df_to_clean:
    variables.loc[var] = [var, df_to_clean[var].dtype, df_to_clean[var].nunique(), df_to_clean[var].unique().tolist()]
    
variables
```

```python
df_to_clean['Annual Salary ($USD)'].str.replace('$','').unique()
```

```python
df_to_clean['Annual Salary ($USD)'] = (
    df_to_clean['Annual Salary ($USD)']
    .str.replace('$','')
    .str.replace(',','')
    .str.replace('>','')
)
df_to_clean['Annual Salary ($USD)'].unique()
```

```python
val1 = df_to_clean['Annual Salary ($USD)'].str.split('-', expand=True)[0].fillna('0').astype(int)
val2 = df_to_clean['Annual Salary ($USD)'].str.split('-', expand=True)[1].fillna('0').astype(int)
```

```python
df_to_clean['Average Annual Salary ($USD)'] = (val1 + val2)/ 2
df_to_clean['Average Annual Salary ($USD)'].iloc[:5]
```

```python
df_to_clean['Company Size'] = df_to_clean['Company Size'].str.replace('employees','')
df_to_clean['Coding Experience'] = df_to_clean['Coding Experience'].str.replace('years','')
df_to_clean.drop('Annual Salary ($USD)', axis=1, inplace=True)
df_to_clean.head()
```

```python
df_clean = df_to_clean.copy()
df_clean.dropna(inplace=True)
```
